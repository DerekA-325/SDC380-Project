```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime

student_id = "derave1577"
current_datetime = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

df_new = pd.read_excel('C:\\Users\\Student\\Downloads\\Stroke_Prediction_Dataset.xlsx')

plt.figure(figsize=(10, 6))
plt.hist(df_new['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.title(f'Histogram of Age Distribution Among Patients - {student_id} - {current_datetime}')
plt.grid(True)
plt.show()
```

```python
stroke_diagnosis = df_new[df_new['Diagnosis'] == 'Stroke']
work_type_counts = stroke_diagnosis['Work Type'].value_counts()

plt.figure(figsize = (10, 6))
plt.barh(work_type_counts.index, work_type_counts.values, color = 'green')
plt.xlabel('Number of Patients')
plt.ylabel('Work Type')
plt.title(f'Work Type Distribution of Patients Diagnosed with a Stroke - {student_id} - {current_datetime}')
plt.grid(True)
plt.show()
```

```python
x = stroke_diagnosis['Age']
y = stroke_diagnosis['Body Mass Index (BMI)']
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.figure(figsize=(12, 8))
plt.scatter(x, y, label='Data Points', color='blue')
plt.plot(x, p(x), label=f'Trendline: y = {z[0]:.2f}x + {z[1]:.2f}', color='red')
plt.xlabel('Age')
plt.ylabel('Body Mass Index (BMI)')
plt.title(f'Age vs Body Mass Index (BMI) with Trendline for Patients Diagnosed with Stroke - {student_id} - {current_datetime}')
plt.legend()
plt.show()
```

```python
selected_columns = ['Age', 'Body Mass Index (BMI)', 'Average Glucose Level']
correlation_matrix = stroke_diagnosis[selected_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title(f'Heatmap of Correlations Between Age, BMI, and AGL - {student_id} - {current_datetime}')
plt.show()
```
