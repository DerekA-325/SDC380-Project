```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, f1_score, precision_score, recall_score

student_id = "derave1577"
file_path = "C:\\Users\\Student\\Downloads\\Stroke_Prediction_Dataset.xlsx"

df = pd.read_excel(file_path)

categorical_cols = ['Gender', 'Marital Status', 'Work Type', 'Residence Type', 'Smoking Status', 'Alcohol Intake', 'Physical Activity', 'Family History of Stroke', 'Dietary Habits', 'Diagnosis']

df_encoded = pd.get_dummies(df, columns=categorical_cols, dtype=int)
display(df_encoded.head())
print(df_encoded.columns)
```

```python
X_class = df_encoded[['Age', 'Work Type_Self-employed', 'Work Type_Government Job', 'Work Type_Private', 'Work Type_Never Worked', 'Body Mass Index (BMI)', 'Average Glucose Level']]
y_class = df_encoded['Diagnosis_Stroke']

X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)
```

```python
linear_model = LinearRegression()
logistic_model = LogisticRegression(random_state=42)

linear_model.fit(X_train_class, y_train_class)
logistic_model.fit(X_train_class, y_train_class)

y_pred_linear = linear_model.predict(X_test_class)
y_pred_logistic = logistic_model.predict(X_test_class)

accuracy = accuracy_score(y_train_class, logistic_model.predict(X_train_class))

X_test_class['Predicted_Logistic'] = y_pred_logistic
accuracy_logistic = accuracy_score(y_test_class, y_pred_logistic)
f1 = f1_score(y_test_class, y_pred_logistic, average='weighted')
precision = precision_score(y_test_class, y_pred_logistic)
recall = recall_score(y_test_class, y_pred_logistic)

mae = mean_absolute_error(y_train_class, linear_model.predict(X_train_class))
mse = mean_squared_error(y_train_class, linear_model.predict(X_train_class))
r2 = r2_score(y_train_class, linear_model.predict(X_train_class))

display(f'Accuracy: {accuracy_logistic}')
display(f'Precision: {precision}')
display(f'Recall: {recall}')
display(f'F1 Score: {f1}')

display(f'MAE: {mae}')
display(f'MSE: {mse}')
display(f'R2 Score: {r2}')
```
